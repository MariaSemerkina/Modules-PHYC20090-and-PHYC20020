# Calculation of MIllikan Oil Drop Exp

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.optimize import curve_fit
%matplotlib inline

# The below commands make the font and image size bigger
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
#we take vf=x and vr=y
c=(9*(1.834e-5))/(2*886*9.8)
b = (8.2e-3)/(2*(101400))
```

```python
def q1(x,y):
    return ((4*np.pi)/3)*((np.sqrt((b)**2+(c*x))-b)**3)*((886*9.8*(7e-3))*(x+y))/(377*x)
    
```

```python
def dqdd(x,y):
    return ((4*np.pi)/3)*((np.sqrt((b)**2+(c*x))-b)**3)*(886*9.8*(x+y))/(377*x)
                                                     
```

```python
def dqdV(x,y):
    return ((4*np.pi)/3)*(((np.sqrt((b)**2+(c*x))-b)**3)*(-(886*9.8*(7e-3)*(x+y)))/((377**2)*x))
    
```

```python
def dqdmu(x,y):
    return ((4*np.pi)/3)*((3*9*x)/(2*886*9.8))*((((np.sqrt((b)**2+(c*x))-b)**2))/((2)*np.sqrt(((b)**2)+(c*x))))*((886*9.8*(7e-3))*(x+y))/(377*x)
    
```

```python
def dqdvf(x,y):
    return ((4*np.pi)/3)*((3*c*x)*((np.sqrt((b)**2+(c*x))-b)**2)*(((886*9.8*(7e-3))*(x+y)))-(2*(886*9.8*(7e-3)*y)*((np.sqrt(((b)**2)+(c*x))-b)**3)*(np.sqrt((b)**2+(c*x)))))/(2*377*(x**2)*(np.sqrt((b)**2+(c*x))))
    
            
            
```

```python
def dqdvr(x,y):
    return ((4*np.pi)/3)*((np.sqrt((b)**2+(c*x))-b)**3)*(886*9.8*(7e-3))/(377*x)
```

```python
def error(x,y):
    return ((dqdd(x,y))**2)*((5e-6)**2)+((dqdV(x,y))**2)*(2)+((dqdvf(x,y))**2)*((4.934e-5)**2)+((dqdvr(x,y))**2)*((1.495e-4)**2)+((dqdmu(x,y))**2)*((.02e-5)**2)


```

```python
def q(x,y):
    return ((4*np.pi)/3)*((np.sqrt((b)**2+(c*x))-b)**3)*((886*9.8*(7.2e-3))*(x+y))/(300*x),'+-',np.sqrt(error(x,y))
```

```python
x1=np.array([9.334e-19, 10.897e-19, 5.383e-19, 9.764e-19, 9.3076e-19, 3.391e-19, 6.8869e-19, 6.3454e-19, 26.697e-19, 3.2476e-19])
xerr=np.array([8.9809e-19, 9.5693e-19, 7.427e-19, 17.4224e-19, 10.2307e-19, 6.2796e-19, 9.0435e-19, 7.7917e-19, 13.174e-19, 6.56864e-19])
n=np.array([6,7,3,6,6,2,4,4,17,2])
nerr = (.5)
```

```python
np.mean(xerr)
```

```python
np.mean(x1)
```

```python
def fitfunc(x,m,c):
    return m*x+c
```

```python
plt.plot(n,x1,'bo')
pars, cov = curve_fit(fitfunc,n,x1)
plt.plot(n,fitfunc(n,*pars),'g-')
plt.errorbar(n,x1,xerr,nerr, fmt='o',capsize=5)

plt.title("Charge against Integer Number of Electrons")
plt.xlabel('Integer Number of Electrons Rounded')
plt.ylabel('Charge of the Oil Drop Measured (C)')
plt.ylim(-.8e-18,4.2e-18)
plt.grid(True)

```
