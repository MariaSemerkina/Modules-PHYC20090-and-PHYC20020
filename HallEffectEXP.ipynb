Data Analysis for Hall Effect


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.optimize import curve_fit
%matplotlib inline
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
def LLS(x,y):
    numer=0
    denom=0
    meanx=np.mean(x)
    meany=np.mean(y)
    for i in range(np.size(x)):
        #numerator of fraction of linear least squares
        numer += (x[i] - meanx)*(y[i] - meany)
        #denominator of fraction calculation
        denom += (x[i] - meanx)**2
    m = numer/denom
    c = meany-m*meanx
    return m,c
#This function calculates the Linear Least Squares Fit to the data using
#the original equations from first principles
```

```python
x1current=np.array([.54,.6,.66,.72,.78,.84,.9,.95,1.04,1.13,1.41,1.53,
                    1.61,1.69,1.8,1.92,2.18,2.96,3.05,3.11,3.72,3.82,
                    4.93,5.03,5.14,5.31,5.38,5.43,7.27,7.43,7.56])
Vhy1=np.array([-5.2,-5.8,-6.3,-6.8,-7.4,-7.9,-8.4,-8.9,-9.7,-10.5,
              -13,-14,-14.7,-15.5,-16.4,-17.5,-19.8,-26.7,-27.6,-28,
              -33.4,-34.3,-44.3,-45.1,-46,-47.5,-48.2,-48.7,-64.7,
              -66,-67])
```

```python
LLS(x1current,Vhy1)
#As you can see here and below the Linear Least Square is the exact same
#result when a python curve_fit function is applied so the python curve
#fit will be used from now on
```

```python
def fitfunc(x,m,c):
    return m*x+c
```

```python
plt.plot(x1current,Vhy1,'bo')
pars, cov = curve_fit(fitfunc,x1current,Vhy1)
plt.plot(x1current,fitfunc(x1current,*pars),'g-')

plt.title("Hall Voltage vs Currrent with a Magnetic Field at -0.0398T")
plt.xlabel('Current (mA)')
plt.ylabel('Hall Voltage (mV)')
plt.grid(True)
```

```python
print(pars[0],'is the slope')
```

```python
print(cov)
```

```python
print('error on the slope is',np.sqrt(cov[0][0]))
```

```python
print('error on the intercept is',np.sqrt(cov[1][1]))
```
